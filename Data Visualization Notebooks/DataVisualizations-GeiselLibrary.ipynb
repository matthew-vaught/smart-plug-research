```python
import pandas as pd
import os
import glob

# Specify the path to your CSV files
path = r'C:\ResearchFiles\AllFiles_Cleaned\Geisel_Library'
all_files = glob.glob(os.path.join(path, "*.csv"))

# Combine all files into a single DataFrame
df_list = []
for filename in all_files:
    df = pd.read_csv(filename)
    df['smart_plug_id'] = os.path.basename(filename).split('_')[1]  # Add a column for the smart plug identifier
    df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)
```

```python
# Drop unneccessary columns
combined_df = combined_df.drop(columns = ['analogInput_2', 'analogInput_4', 'analogInput_5', 'binaryInput_3', 'binaryValue_1'])

# Convert timestamp column to datetime
combined_df['time'] = pd.to_datetime(combined_df['time'])

# Set timestamp as index and rename the poewr consumption column for better readability
combined_df.set_index('time', inplace=True)
combined_df = combined_df.rename(columns = {'analogInput_3': 'power_consumption (mW)'})
```

```python
combined_df
```

```python
# Forward fill any missing values
combined_df.fillna(method='ffill', inplace=True)
```

```python
# Resample to daily intervals for easier graphing abilities
daily_df = combined_df.resample('D').mean(numeric_only=True)
```

```python
daily_df
```

```python
# Creating a graph that overlays winter break with the time series data to show the dip in power consumption during this period

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define the date range to show in the plot
start_date = '2023-11-01'
end_date = '2024-02-29'

# Define the winter break date range
winter_break_start = '2023-12-18'
winter_break_end = '2024-01-01'

# Plotting overall trends
plt.figure(figsize=(14, 7))
ax = sns.lineplot(data=daily_df, x=daily_df.index, y='power_consumption (mW)')
plt.title('Daily Average Power Consumption')
plt.xlabel('Date')
plt.ylabel('Power')

# Highlight the winter break period with a different color (e.g., light blue)
plt.axvspan(pd.to_datetime(winter_break_start), pd.to_datetime(winter_break_end), color='red', alpha=0.3, label='Winter Break')

# Set the x-axis limits
plt.xlim(pd.to_datetime(start_date), pd.to_datetime(end_date))

# Show the legend
plt.legend()

plt.show()
```

```python
# Creating a heatmap with day of the month on the y-axis and month on the x-axis to show daily power consumption

# Extract month and day information
daily_df['month'] = daily_df.index.month
daily_df['day'] = daily_df.index.day

# Create a pivot table with months on the x-axis and days on the y-axis
pivot_table = daily_df.pivot_table(values='power_consumption (mW)', index='day', columns='month', aggfunc='mean')

# Define the order of the months for better visualization
month_order = ['Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar', 'Apr', 'May']
month_mapping = {10: 'Oct', 11: 'Nov', 12: 'Dec', 1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May'}
pivot_table.columns = pivot_table.columns.map(month_mapping)

# Select only the months of interest
pivot_table = pivot_table[['Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar', 'Apr', 'May']]

# Create the heatmap
plt.figure(figsize=(14, 7))
sns.heatmap(pivot_table, cmap="YlGnBu", cbar_kws={'label': 'Average Power Consumption'}, linewidths=0, linecolor='gray')

# Customize the heatmap for better presentation
plt.title('Daily Average Power Consumption by Month')
plt.xlabel('Month')
plt.ylabel('Day of Month')
plt.show()
```
